For this problem we use a sliding window/two pointer technique; we have two variables, L and R, which are both initialized as position 0. The question asks for number of combinations of soldiers with a specified max difference in height. This can be completed with an approach of nested for loops, but that is not very efficient. So instead, the algorithm starts L (left) and R (right) at position 0, and then it moves the right pointer one position to the right if it is in the specified height range. This continues until we hit the end of the list.

```python
n, d = map(int, input().split())
a = sorted(map(int, input().split()))

ans = 0
R = 0

for L in range(n):
    while R + 1 < n and a[R + 1] - a[L] <= d:
        R += 1

    ans += 2 * (R - L)

    if R == L:
        R += 1

print(ans)
```
